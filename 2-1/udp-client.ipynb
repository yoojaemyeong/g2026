## UDP 소켓 :: 클라이언트편

```
 UDP 서버
─────────────────────  
socket()  
bind()

sendto()
close()

========================================================

 UDP 클라이언트
─────────────────────  
socket()               
sendto()     ← 바로 전송
recvfrom()
close()
```

---

### UDP 클라이언트
```python
import socket

# 1. 소켓 생성
client = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)

# 2. 데이터 송신 (connect 없이 바로 전송!)
client.sendto("UDP 메시지".encode(), ("127.0.0.1", 9999))

# 3. 응답 수신
data, addr = client.recvfrom(1024)
print(f"서버 응답: {data.decode()}")

# 4. 종료
client.close()
```

````python
# UDP 클라이언트
import socket

# 1. 소켓 생성
client = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)

# 2. 데이터 송신 (connect 없이 바로 전송!)
client.sendto("UDP 메시지".encode(), ("127.0.0.1", 9999))

# 3. 응답 수신
data, addr = client.recvfrom(1024)
print(f"서버 응답: {data.decode()}")

# 4. 종료
client.close()
````
