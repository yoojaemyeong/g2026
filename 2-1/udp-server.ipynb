## UDP 소켓 :: 서버편

```
 UDP 서버
─────────────────────  
socket()  
bind()    
recvfrom()   ← 바로 수신
sendto()
close()

========================================================

 UDP 클라이언트
─────────────────────  
socket()               
sendto()     ← 바로 전송
recvfrom()
close()
```

---

### UDP 서버
```python
import socket

# 1. 소켓 생성 (SOCK_DGRAM)
server = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)

# 2. 주소 바인딩
server.bind(("0.0.0.0", 9999))
print("UDP 서버 대기 중...")

# 3. 데이터 수신 (연결 수락 없음!)
data, addr = server.recvfrom(1024)  # 데이터 + 클라이언트 주소 반환
print(f"수신: {data.decode()} / 주소: {addr}")

# 4. 응답 송신
server.sendto("UDP 응답".encode(), addr)

# 5. 종료
server.close()
```

````python
# UDP서버
import socket

# 1. 소켓 생성 (SOCK_DGRAM)
server = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)

# 2. 주소 바인딩
server.bind(("0.0.0.0", 9999))
print("UDP 서버 대기 중...")

# 3. 데이터 수신 (연결 수락 없음!)
data, addr = server.recvfrom(1024)  # 데이터 + 클라이언트 주소 반환
print(f"수신: {data.decode()} / 주소: {addr}")

# 4. 응답 송신
server.sendto("UDP 응답".encode(), addr)

# 5. 종료
server.close()
````
