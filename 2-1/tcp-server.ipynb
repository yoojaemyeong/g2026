```
 TCP 서버                      
─────────────────────         
socket()                      
bind()

recv()

close()

=============================

 TCP 클라이언트               
─────────────────────        
socket()                     
connect()  ← 연결 필수        
send()                       
recv()
close()
```

---

### TCP 서버
```python
import socket

# 1. 소켓 생성
server = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

# 2. 주소 재사용 옵션 (재시작 시 포트 충돌 방지)
server.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR, 1)

# 3. 주소 바인딩
server.bind(("0.0.0.0", 9999))

# 4. 연결 대기 (최대 5개 클라이언트 큐)
server.listen(5)
print("서버 대기 중...")

# 5. 클라이언트 연결 수락
conn, addr = server.accept()
print(f"연결됨: {addr}")

# 6. 데이터 수신 / 송신
data = conn.recv(1024)          # 최대 1024 bytes 수신
print(f"수신: {data.decode()}")
conn.send("응답 완료".encode())  # 데이터 송신

# 7. 종료
conn.close()
server.close()
```

````python
# TCP서버
import socket

# 1. 소켓 생성
server = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

# 2. 주소 재사용 옵션 (재시작 시 포트 충돌 방지)
server.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR, 1)

# 3. 주소 바인딩
server.bind(("0.0.0.0", 9999))

# 4. 연결 대기 (최대 5개 클라이언트 큐)
server.listen(5)
print("서버 대기 중...")

# 5. 클라이언트 연결 수락
conn, addr = server.accept()
print(f"연결됨: {addr}")

# 6. 데이터 수신 / 송신
data = conn.recv(1024)          # 최대 1024 bytes 수신
print(f"수신: {data.decode()}")
conn.send("응답 완료".encode())  # 데이터 송신

# 7. 종료
conn.close()
server.close()
````
