## TCP 소켓 :: 클라이언트편

```
 TCP 서버                      
─────────────────────         
socket()                      
bind()                        
listen()                      
accept()   ← 연결 대기        
recv()
send()
close()

=============================

 TCP 클라이언트               
─────────────────────        
socket()                     
connect()  ← 연결 필수        
send()                       
recv()
close()
```

---

### TCP 클라이언트
```python
import socket

# 1. 소켓 생성
client = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

# 2. 서버에 연결
client.connect(("127.0.0.1", 9999))

# 3. 데이터 송신 / 수신
client.send("안녕하세요".encode())
response = client.recv(1024)
print(f"서버 응답: {response.decode()}")

# 4. 종료
client.close()
```

````python
# TCP 클라이언트
import socket

# 1. 소켓 생성
client = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

# 2. 서버에 연결
client.connect(("127.0.0.1", 9999))

# 3. 데이터 송신 / 수신
client.send("안녕하세요".encode())
response = client.recv(1024)
print(f"서버 응답: {response.decode()}")

# 4. 종료
client.close()
````
