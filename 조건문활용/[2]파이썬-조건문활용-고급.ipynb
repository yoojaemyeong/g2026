* 드론에서 HUD(Head-Up Display)란 조종사(사용자)가 시선을 돌리지 않고도 전방 시야에서 실시간 비행 정보를 확인할 수 있도록 화면에 띄워주는 그래픽 인터페이스를 뜻함
* 조종기와 연결된 스마트폰이나 FPV 고글 화면에 주로 표시됨

## Level 1(하) np.where 단일 조건 

```python
# Level 1(하) np.where 단일 조건 엔진 가동
import numpy as np

#드론 센서 데이터: 고도(m)
altitude_data = np.array([5,68,12,15, 7])

#  np.where 적용: 10 초과면 '경고' 아니면 '안전'
hud_warning = np.where(altitude_data > 10, '경과(Warning)', '안전(Safe)')

print(hud_warning)
```

## Level 2(중) : 논리 연산자 ( & , | ) : 다중 센서 데이터 제어

```python
# 복합 센서 안전 필터링 시스템
# Code Injection: 애플리케이션의 보안 취약점을 악용하여, 공격자가 시스템에 악의적인 코드나 명령어를 주입하고 실행시키는 사이버 공격 기법
import numpy as np

altitude = np.array([8,12, 5,15]) #고도
wind_speed = np.array([3,2,6,4])  #풍속

#다중 조건 병합( & 사용)
flight_status = np.where(
    (altitude <= 10) & (wind_speed < 5), 
    '비행 승인(Clear)',
    '비행 금지(No Fly)'
)

print(flight_status)
```

## Level 3(상) : 실시간 다중 색상 배터리 HUD(HeadUp Display) 구축

<font color=red> np.select(conditions, choices, default) </font>

```python
# 다이내믹 배터리 디스플레이 시스템
import numpy as np

#실시간 배터리 잔량 데이터 (%)
battery_levels  = np.array([95, 65, 80, 25, 45])

#1. 조건 리스트 생성 (순서 중요)
conditions = [
    (battery_levels >= 70),
    (battery_levels >= 40)
]

#2. 선택 리스트 생성(조건 리스트와 1:1 매칭)
choices = ['Neon_Green', 'Yellow']

#3. np.select 엔진 가동 (그 외는 모두 Red)
hud_colors  =np.select(conditions, choices, default='Crimson_Red')

print(hud_colors)
```
