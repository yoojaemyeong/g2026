# 파이썬 조건문: 컴퓨터에게 생각하는 법 가르치기

## 1. True(참) vs. False(거짓)

```python
# 양수/음수 판별식 만들기
number = 10

if number > 0:
    print(True)
    print("양수입니다.")
else:
    print(False)
    print("양수가 아닙니다.")
```

```python
# 홀수/짝수 판별식 만들기
number = 7

if number % 2 == 0:
    print(True)
    print("짝수입니다.")
else:
    print(False)
    print("홀수입니다.")
```

## 2. 데이터 확인 가격

```python
# 문자열 비교
password = "1234"

if password == "1234":
    print(True)
    print("로그인 성공")
else:
    print(False)
    print("비밀번호 오류")
```

```python
# 놀이기구 탑승하기
height = 140
age = 10

if height >= 130 and age >= 8:
    print(True)
    print("탑승 가능합니다.")
else:
    print(False)
    print("탑승할 수 없습니다.")
```

```python
# 사용자 아이디/패스워드 정보 확인
user_id = "admin"
password = "1234"
is_admin = True

if user_id == "admin" and password == "1234":
    
    if is_admin:
        print(True)
        print("관리자 로그인 성공")
    else:
        print(True)
        print("일반 사용자 로그인 성공")

else:
    print(False)
    print("로그인 실패")
```

## 게임 형식의 영역 채우기 프로그램 작성
<img src=attachment:69b86f52-9d3a-4641-9ef9-9d5a44053de0.png width=800 height=600>

```python
# 코드작성
import turtle

turtle.shape("turtle")
turtle.penup()

while True:
    x = input("X위치:")
    if x == '':
        print("프로그램을 종료합니다.")
        break
    else:
        x = int(x)
    y = int(input("Y위치:"))
    text = input("영역에 표시할 메시지 입력:")

    turtle.goto(x,y)
    turtle.write(text, font=("Arial", 30))
    
turtle.exitonclick()
```

# 여러 조건 : 오늘의 날씨는? ( if... elif ... elif ...else )

```python
#성적 처리 프로그램
score = 85

if score >= 90:
    print("A 학점")
    
elif score >= 80:
    print("B 학점")
    
elif score >= 70:
    print("C 학점")
    
else:
    print("재시험")
```

```python
#봄, 여름, 가을, 겨울 ??
month = 1

if month == 12 or month == 1 or month == 2:
    print("겨울")

elif month >= 3 and month <= 5:
    print("봄")

elif month >= 6 and month <= 8:
    print("여름")

elif month >= 9 and month <= 11:
    print("가을")

else:
    print("잘못된 입력")
```

# 논리 연산자 매트릭스 : 복합 추론의 시작

```python
# ATM출금 시스템 구현
balance = 50000
password = "1234"
input_password = "1234"
withdraw = 30000

if input_password != password:
    print("비밀번호 오류")

elif withdraw > balance:
    print("잔액 부족")

elif withdraw <= 0:
    print("잘못된 금액")

else:
    balance = balance - withdraw
    
    print("출금 완료")
    print("남은 잔액:", balance)
```

```python
# ------------------------
# Level 1 : 기본 조건문 (크루즈 컨트롤)
# ------------------------
speed = 50
target_speed = 60

if speed < target_speed:
    print("가속합니다.")
else:
    print("현재 속도 유지")
```

```python
# ------------------------
# Level 3 : 복합 조건문 (조건부 자동화)
# ------------------------
distance = 8         # 앞차와 거리
safe_gap = 10        # 안전 거리
lane_clear = True    # 옆 차선이 비어있는지
obstacle = False     # 장애물 여부

if distance < safe_gap and lane_clear:
    print("차선을 변경합니다.")
elif obstacle:
    print("브레이크를 작동합니다.")
else:
    print("속도 유지")
```

```python
# ------------------------
# Level 5 : 고급 조건문 (완전 자율주행 - 단순화 버전)
# ------------------------
distance = 5
lane_clear = False
pedestrian = True
traffic_light = "red"

if pedestrian:
    print("보행자 발견: 정지")
elif traffic_light == "red":
    print("신호등 빨간불: 정지")
elif distance < safe_gap and lane_clear:
    print("차선 변경")
elif distance < safe_gap:
    print("감속")
else:
    print("정상 주행")
```

```python
# 드론 상태 및 환경 정보
battery = 25       # 배터리 잔량 (%)
wind_speed = 8     # 풍속 (m/s)
gps_signal = True  # GPS 신호 수신 여부

# 조건 판단 (Logic Engine)
if battery > 20 and wind_speed < 10 and gps_signal:
    print("비행 가능: 드론이 안정적으로 비행합니다.")
elif battery <= 20:
    print("경고: 배터리가 부족합니다.")
elif wind_speed >= 10:
    print("주의: 바람이 강하여 비행이 어렵습니다.")
elif not gps_signal:
    print("위험: GPS 신호 없음 - 비행 불가")
else:
    print("비행 불가: 안전 조건 미충족")
```

# 게임 속 보이지 않는 규칙 설계
<img src=attachment:3bd8b7fb-274a-48df-8f44-1f9ec7a37a7d.png width=800 height=600>

```python
import random

# 상태 설정
monster_hp = 15          # 몬스터 체력 (퍼센트)
player_state = "normal"  # "normal", "stunned" (기절)

# 몬스터 행동 결정
if monster_hp < 20:
    print("몬스터 행동: [회복 스킬 사용]")
elif player_state == "stunned":
    print("몬스터 행동: [강력 공격]")
else:
    # 랜덤 일반 공격 선택
    attacks = ["물기 공격", "발톱 공격", "돌진 공격"]
    selected_attack = random.choice(attacks)
    print(f"몬스터 행동: [일반 공격 - {selected_attack}]")
```
